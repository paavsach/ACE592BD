```python
page = requests.get("https://www.epa.gov/criteria-air-pollutants/naaqs-table")
soup = BeautifulSoup(page.content, 'html.parser')
data = [[cell.get_text(strip=True) for cell in row.find_all('td')] for row in soup.find_all("tr")]
for entry in data:
    print(str(entry))
```

```python
import os
txtfile = open(os.path.join(os.path.expanduser('~'), "outputs/Lab1", "epa_norm_scraper.txt"),"w")
txtfile.write("Pollutant, Primary/Secondary, Avg_Time, Level, Form")
for entry in data:
    text = str(entry)
    text = text.replace(']', '')
    text = text.replace('[', '')
    text = text.replace("'", '')
    print(text)
    txtfile.write(text + "\n")
    
txtfile.close()

```
